```python
start.rename(columns={'R&D Spend':'RDspend','Marketing Spend':'Marketingspend'},inplace=True)
start.head(5)
```

```python
import statsmodels.formula.api as smf
```

```python
model= smf.ols('Profit~RDspend+Administration+Marketingspend',data= start).fit()
```

```python
model.params
```

```python
print(model.tvalues,"\n",model.pvalues)
```

```python
(model.rsquared,model.rsquared_adj)
```

```python
mod_admin= smf.ols('Profit~Administration',data=start).fit()
print(mod_admin.tvalues,'/n',mod_admin.pvalues)
```

```python
mod_mark= smf.ols('Profit~Marketingspend',data=start).fit()
print(mod_mark.tvalues,"/n",mod_mark.pvalues)
```

```python
mod_adm_mark=smf.ols('Profit~Marketingspend+Administration',data=start).fit()
print(mod_adm_mark.tvalues,"/n",mod_adm_mark.pvalues)
```

```python
rsq_rd= smf.ols('RDspend~Administration+Marketingspend',data=start).fit().rsquared
vif_rd=1/(1-rsq_rd)#variance inflation factor
rsq_Admi= smf.ols('Administration~RDspend+Marketingspend',data=start).fit().rsquared
vif_admi=1/(1-rsq_Admi)
rsq_MS= smf.ols('Marketingspend~Administration+RDspend',data=start).fit().rsquared
vif_ms=1/(1-rsq_MS)
```

```python
set={'variables':['RDspend','Administration','Marketingspend'],'vif_values':[vif_rd,vif_admi,vif_ms]}
vif = pd.DataFrame(set)
vif
```

QQ Plot

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q')
plt.title("normal qq plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-20000))
```

```python
def get_standardized_values(vals):
    return (vals-vals.mean())-vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),get_standardized_values(model.resid))
plt.title("residual plot")
plt.xlabel('standardized fitted values')
plt.ylabel('standardized residual values')
plt.show()
```

```python
figure = plt.figure(figsize=(16,9))
figure = sm.graphics.plot_regress_exog(model,"RDspend", fig=figure)
plt.show()
```

```python
figure = plt.figure(figsize=(16,9))
figure = sm.graphics.plot_regress_exog(model,"Administration", fig=figure)
plt.show()
```

```python
figure = plt.figure(figsize=(16,9))
figure = sm.graphics.plot_regress_exog(model,"Marketingspend", fig=figure)
plt.show()
```

```python
model_influence = model.get_influence()
(c,_)=model_influence.cooks_distance
```

```python
fig = plt.subplots(figsize=(22,8))
plt.stem(np.arange(len(start)),np.round(c,3))
plt.xlabel("rows index")
plt.ylabel("cooks distance")
plt.show()
```

```python
(np.argmax(c),np.max(c))
```

```python
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()
```
