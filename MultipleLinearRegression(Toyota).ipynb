Model Building

```python
import statsmodels.formula.api as smf
```

```python
model= smf.ols('price~age+km+hp+cc+doors+gears+quaterly_tax+weight',data= corola).fit()
```

```python
model.summary()
```

```python
model.params
```

```python
print(model.tvalues,"\n", model.pvalues)
```

```python
(model.rsquared,model.rsquared_adj)
```

```python
mod_cc= smf.ols('price~cc',data=corola).fit()
print(mod_cc.tvalues,'/n',mod_cc.pvalues)
```

```python
mod_doors= smf.ols('price~doors',data=corola).fit()
print(mod_doors.tvalues,'/n',mod_doors.pvalues)
```

```python
mod_cc_doors= smf.ols('price~cc+doors',data=corola).fit()
print(mod_cc_doors.tvalues,'/n',mod_cc_doors.pvalues)
```

Model Validation

```python
rsq_age= smf.ols('age~km+hp+cc+doors+gears+quaterly_tax+weight',data=corola).fit().rsquared
vif_age=1/(1-rsq_age)
rsq_km= smf.ols('km~age+hp+cc+doors+gears+quaterly_tax+weight',data=corola).fit().rsquared
vif_km=1/(1-rsq_km)
rsq_hp= smf.ols('hp~age+age+cc+doors+gears+quaterly_tax+weight',data=corola).fit().rsquared
vif_hp=1/(1-rsq_hp)
rsq_cc= smf.ols('cc~km+hp+age+doors+gears+quaterly_tax+weight',data=corola).fit().rsquared
vif_cc=1/(1-rsq_cc)
rsq_doors= smf.ols('doors~km+hp+cc+age+gears+quaterly_tax+weight',data=corola).fit().rsquared
vif_doors=1/(1-rsq_doors)
rsq_gears= smf.ols('gears~km+hp+cc+doors+age+quaterly_tax+weight',data=corola).fit().rsquared
vif_gears=1/(1-rsq_gears) 
rsq_quaterly_tax= smf.ols('quaterly_tax~km+hp+cc+doors+gears+age+weight',data=corola).fit().rsquared
vif_quaterly_tax=1/(1-rsq_quaterly_tax)
rsq_weight= smf.ols('weight~km+hp+cc+doors+gears+quaterly_tax+age',data=corola).fit().rsquared
vif_weight=1/(1-rsq_weight) 
```

```python
set={'variables':['age','km','hp','cc','doors','gears','quaterly_tax','weight'],'vif_values':[vif_age,vif_km,vif_hp,vif_cc,vif_doors,vif_gears,vif_quaterly_tax,vif_weight]}
vif = pd.DataFrame(set)
vif
```

qqplot

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q')
plt.title("normal qq plot of residuals")
plt.show()
```

```python
list(np.where(model.resid>5000))
```

```python
list(np.where(model.resid<-6000))
```

```python
def get_standardized_values(vals):
    return (vals-vals.mean())-vals.std()
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),get_standardized_values(model.resid))
plt.title("residual plot")
plt.xlabel('standardized fitted values')
plt.ylabel('standardized residual values')
plt.show()
```

Checking Outliers

```python
model_influence = model.get_influence()
(c,_)=model_influence.cooks_distance
```

```python
fig = plt.subplots(figsize=(10,12))
plt.stem(np.arange(len(corola)),np.round(c,3))
plt.xlabel("rows index")
plt.ylabel("cooks distance")
plt.show()
```

```python
(np.argmax(c),np.max(c))
```
